```python
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(by=["player", "date"])
```

```python
df.columns
```

```python
df = df.rename(columns={
    "striker": "player",
    "runs_off_bat": "runs",
    "venue_x": "venue",
    "bowling_team": "opponent"
})
```

```python
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(by=["player", "date"])
```

```python
df[["player", "runs", "venue", "opponent", "date"]].head()
```

```python
df["runs_last_5"] = (
    df.groupby("player")["runs"]
      .rolling(5, min_periods=1)
      .mean()
      .reset_index(level=0, drop=True)
)

df["runs_last_10"] = (
    df.groupby("player")["runs"]
      .rolling(10, min_periods=1)
      .mean()
      .reset_index(level=0, drop=True)
)
```

```python
df["venue_avg_runs"] = (
    df.groupby(["player", "venue"])["runs"]
      .transform("mean")
)
```

```python
df["opponent_avg_runs"] = (
    df.groupby(["player", "opponent"])["runs"]
      .transform("mean")
)
```

```python
df["career_avg_runs"] = (
    df.groupby("player")["runs"]
      .transform("mean")
)
```

```python
df.fillna({
    "runs_last_5": df["career_avg_runs"],
    "runs_last_10": df["career_avg_runs"],
    "venue_avg_runs": df["career_avg_runs"],
    "opponent_avg_runs": df["career_avg_runs"]
}, inplace=True)
```

```python
df.to_csv("../data/features/ipl_features.csv", index=False)
```
