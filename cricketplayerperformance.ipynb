```python
#data preprocessing
#check missing value
print(ball.isnull().sum())
print(matches.isnull().sum())
```

```python
#Fill Now values to null dataframe
ball.fillna(0,inplace=True)
```

```python
#data cleaning
#Separate numeric and object columns in matches dataframe
num_cols = matches.select_dtypes(include=['int64','float64']).columns
obj_cols = matches.select_dtypes(include=['object']).columns
```

```python
 matches[num_cols]=matches[num_cols].fillna(0)
 matches[obj_cols]=matches[obj_cols].fillna("Unknown")
```

```python
print(matches.isnull().sum())
```

```python
#Standardize column names
ball.columns = ball.columns.str.lower().str.replace(" ","_")
matches.columns = matches.columns.str.lower().str.replace(" ","_")
```

```python
#convert data column 
matches['date'] =pd.to_datetime(matches['date'], errors='coerce')
print(matches['date'].head())
```

```python
print(ball.columns)
print(matches.columns)
```

```python
#Rename id column to match merge key 
matches.rename(columns={'id': 'match_id'}, inplace=True)
#Merge ball&matches
ipl =ball.merge(matches, on='match_id',how='left')
print(ipl.shape)
```

```python
print(ipl)
```

```python
import matplotlib.pyplot as plt
```

```python
#visualization
plt.figure()
ipl['batsman_runs'].hist(bins=20)
plt.title("Distribution of batsman Runs")
plt.xlabel("Runs per Ball")
plt.ylabel("Frequency")
plt.show()
```

```python
#wicket distribution
print("Wicket distribution")
print(ipl['is_wicket'].value_counts())

plt.figure()
ipl['is_wicket'].value_counts().plot(kind="bar")
plt.title("wicket distribution (0=not out,1=out)")
plt.xlabel("is Wicket")
plt.ylabel("count")
plt.show()
```

```python
#venue status - Average runs 
top_venues = (
    ipl.groupby('venue')['batsman_runs']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure()
top_venues.plot(kind='barh')
plt.title("Top Venues by Average Runs")
plt.xlabel("Venue")
plt.ylabel("Average Runs per Ball")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#team performance and total runs 
top_teams = (
    ipl.groupby('batting_team')['batsman_runs']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure()
top_teams.plot(kind='barh')
plt.title("Top Teams by Total Runs")
plt.xlabel("Total Runs")
plt.ylabel("Team")
plt.show()
```
